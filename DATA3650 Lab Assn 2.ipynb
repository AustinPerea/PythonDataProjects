#### <font color='red'>Que 1:</font> If the basic salary is absent, remove the employee's record from the dataframe. S ubstitute the missing value with the employee's ID if the name of an employee is missing. If an employee's job title is absent, the most common job title should be used in its place. <font color='blue'>(2 point)

```python
df['JobTitle'] = df['JobTitle'].fillna(df['JobTitle'].mode()[0])
df['EmployeeName'] = df['EmployeeName'].fillna(df['Id'])
df
```

#### <font color='red'>Que 2:</font> The total amount of compensation is the sum of the basic wage and any additional compensation received through overtime or other sources. The sum of the whole payment plus the perks is the total amount of payment with benefits. For all employees, update the total amount paid and the total amount paid including benefits. In the circumstances of overtime pay and benefits, substitute the missing values (if any) with 0. Fill out the missing values (if any) of basic payment and other payment forms using the average and median values, respectively.  <font color='blue'>(10 point)

```python
df['Benefits'] = df['Benefits'].fillna(0)
df['OvertimePay'] = df['OvertimePay'].fillna(0)
df['BasePay'] = df['BasePay'].fillna(df['BasePay'].mean())
df['OtherPay'] = df['OtherPay'].fillna(df['OtherPay'].mean())
df
```

#### <font color='red'>Que 3:</font> Find the IDs and Names all employees of the Fire Department. <font color='blue'>(3 point)

```python
df['JobTitle'] = df['JobTitle'].str.upper()
df[df['JobTitle'].str.contains('FIRE')][['EmployeeName', 'Id']]
```
