```python
# libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### <font color='red'> Que 1: </font> Load penguins dataset from seaborn library to a DataFrame. Answer the rest of the questions from this dataset.  <font color='blue'>(1 point)

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

#### <font color='red'> Que 2: </font> (A) Delete an entire record if three or more attributes of the record contains missing values (B) The most frequent 'sex' value of the related species should be used to fill the missing values of the 'sex' attribute.<font color='blue'>(3 point)

```python
penguins.dropna(thresh=3, inplace=True)
penguins.fillna(penguins['sex'].mode()[0], inplace=True)
penguins.head()
```

#### <font color='red'> Que 3: </font> Display two plots side by side. You must include appropriate labels for axis and title of your data visualization. (A) Which species is the most representative group? (B) Show the distribution of the island. <font color='blue'>(4 point)

```python
fig, axes = plt.subplots(1, 2, figsize=(10,5))

# Representative
sns.countplot(ax=axes[0], x=penguins['species'])
axes[0].set_title('Most Representative group')

# Distribution
sns.countplot(ax=axes[1], x=penguins['island'])
axes[1].set_title('Distrubution of the Penguin Islands')
```

#### <font color='red'> Que 4: </font> (A) Write the code to generate the following distribution. Your graph may have different color. (B) What conclusion can you draw from this figure? Create a new cell and write your answer as comment or write your answer in a markdown cell. <font color='blue'>(4 point)

```python
sns.set_theme(style="ticks")

ax = sns.jointplot(x=penguins['flipper_length_mm'], y=penguins['body_mass_g'], kind="hex", color="#BB5F5F")

plt.xlabel('Flipper length measured in mm')
plt.ylabel('Body mass measured in gram')
plt.show()
```

#### <font color='red'> Que 5: </font> For each species, visualize the possible outliers with respect to the flipper length. In data visualisation, take into account the gender of the penguins. You must include appropriate labels for axis, legend and title of your data visualization.<font color='blue'>(3 point)

```python
adelie = penguins[penguins['species'] == 'Adelie']
adGen = adelie['sex']
gentoo = penguins[penguins['species'] == 'Gentoo']
geGen = gentoo['sex']
chinstrap = penguins[penguins['species'] == 'Chinstrap']
chGen = chinstrap['sex']

fig, axes = plt.subplots(1, 3, figsize=(15,5))

sns.boxplot(ax=axes[0], x=adGen, y=penguins['flipper_length_mm'])
axes[0].set_title('Adelie')
sns.boxplot(ax=axes[1], x=geGen, y=penguins['flipper_length_mm'])
axes[1].set_title('Gentoo')
sns.boxplot(ax=axes[2], x=chGen, y=penguins['flipper_length_mm'])
axes[2].set_title('Chinstrap')

plt.show()
```

#### <font color='red'> Que 6: </font> For each species, demonstrate how bodily mass affects each of the following: bill length, bill depth, and flipper length. In data visualisation, take into account the gender of the penguins. For each species, there will be a separate subplot in the figure. You must include appropriate labels for axis and legend in the subplots. Add a common title of the figures.<font color='blue'>(5 point)

```python

fig, axes = plt.subplots(3, 3, figsize=(20,20))
fig.suptitle('How bodily mass affects the bill length, depth, and flipper length')

axes[0,1].set_title('Adelie')
sns.scatterplot(ax=axes[0,0], x=adelie['body_mass_g'], y=penguins['bill_length_mm'], hue=penguins['sex'])
sns.scatterplot(ax=axes[0,1], x=adelie['body_mass_g'], y=penguins['bill_depth_mm'], hue=penguins['sex'])
sns.scatterplot(ax=axes[0,2], x=adelie['body_mass_g'], y=penguins['flipper_length_mm'], hue=penguins['sex'])

axes[1,1].set_title('Gentoo')
sns.scatterplot(ax=axes[1,0], x=gentoo['body_mass_g'], y=penguins['bill_length_mm'], hue=penguins['sex'])
sns.scatterplot(ax=axes[1,1], x=gentoo['body_mass_g'], y=penguins['bill_depth_mm'], hue=penguins['sex'])
sns.scatterplot(ax=axes[1,2], x=gentoo['body_mass_g'], y=penguins['flipper_length_mm'], hue=penguins['sex'])

axes[2,1].set_title('Chinstrap')
sns.scatterplot(ax=axes[2,0], x=chinstrap['body_mass_g'], y=penguins['bill_length_mm'], hue=penguins['sex'])
sns.scatterplot(ax=axes[2,1], x=chinstrap['body_mass_g'], y=penguins['bill_depth_mm'], hue=penguins['sex'])
sns.scatterplot(ax=axes[2,2], x=chinstrap['body_mass_g'], y=penguins['flipper_length_mm'], hue=penguins['sex'])

plt.show()
```
