```python
import pandas as pd
import numpy as np
```

```python
amazon = pd.read_csv('AMAZON.csv', usecols=['Date', 'Open', 'Close'])
apple = pd.read_csv('APPLE.csv', usecols=['Date', 'Open', 'Close'])
microsoft = pd.read_csv('MICROSOFT.csv', usecols=['Date', 'Open', 'Close'])
google = pd.read_csv('GOOGLE.csv', usecols=['Date', 'Open', 'Close'])
```

```python
amazon['Date'] = pd.to_datetime(amazon['Date'])
apple['Date'] = pd.to_datetime(apple['Date'])
microsoft['Date'] = pd.to_datetime(microsoft['Date'])
google['Date'] = pd.to_datetime(google['Date'])
```

```python
amazon.head()
```

```python
amazon.set_index('Date', inplace=True)
apple.set_index('Date', inplace=True)
microsoft.set_index('Date', inplace=True)
google.set_index('Date', inplace=True)
```

```python
company = 'Apple Amazon Google Microsoft'.split()
company
```

```python
df_stocks = pd.concat([apple, amazon, google, microsoft], axis=1, keys=company)
df_stocks.head()
```

```python
# Check for missing values
df_stocks.isnull().sum()
```

```python
print(df_stocks.dtypes)
```

```python
df_stocks['Google']['Close'] = df_stocks['Google']['Close'].apply(lambda x: float(x.split()[0].replace('$', '')))
```

```python
# Fill in missing values
# Google Open has missing values, fill the value of the 'Close' column of previous business day.
df_stocks['Google']['Open'].fillna(df_stocks['Google']['Close'].shift(1))
```

#### <font color='maroon'> Que 6: (A) Find the rate of change on the closing price for each company. Calculate the rate of change as: (old_value - new_value)/old_value. <br> (B) Which company's closing price fluctuated the most in the previous year (2022)?</font> *(Point: 4)*

```python
df_rate = pd.DataFrame()
```

```python
for comp in company:
    df_rate[comp + ' Current'] = df_stocks[comp]['Close']
    df_rate[comp + ' Previous'] = df_stocks[comp]['Close'].shift(1)
```

```python
df_rate
```

```python
df_difference = pd.DataFrame()
```

```python
for comp in company:
    df_difference['Difference'] = df_rate[comp + ' Previous'] - df_rate[comp + ' Current']
df_difference
```
